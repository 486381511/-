```python
from ultralytics import YOLO
import cv2
import time
import numpy as np
```

```python
path1 = r"E:\DSX\UnderwaterFishDetection\runs\detect\train10\weights\best.pt"
model = YOLO(path1) 
```

```python
index_to_id = model.names
print(index_to_id)
```

```python
# 鱼的种类编号 -> 中文名称映射
fish_classes =index_to_id
```

```python
# 选择测试视频
path2 = r"E:\DSX\UnderwaterFishDetection\fish2.mp4"
cap = cv2.VideoCapture(path2)

# 获取视频参数
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 定义输出视频路径和编码器
out = cv2.VideoWriter("output.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))
while (cap.isOpened()):
    
    ret, frame = cap.read()
    if ret == True:

        results = model(frame, stream=True)

        for result in results:

            boxes = result.boxes.xyxy.cpu().numpy()
            class_ = result.boxes.cls.cpu().numpy()
            conf = result.boxes.conf.cpu().numpy()

            for i in range(len(boxes)):

                if conf[i] >= 0.7:

                    cls_name = index_to_id[class_[i]]
                    
                    x1,y1,x2,y2 = boxes[i]

                    mid_x = int((x1+x2)/2)
                    mid_y = int((y1+y2)/2)
                    cv2.circle(frame, (mid_x, mid_y), 5, (0, 255, 0), -1)
                    
                    cv2.rectangle(frame, pt1=(int(x1),int(y1)),pt2=(int(x2),int(y2)),color=(0, 255, 0),thickness=3)
                    cv2.putText(frame, cls_name+" : "+str("{:.2f}".format(conf[i])), (int(x1),int(y1)),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0,255,0),3)

         # 显示并写入视频
        cv2.imshow("Tracking", frame)
        out.write(frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break


cap.release()
cv2.destroyAllWindows()
```
